```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-1,1,50)
y1 = 2*x+1
y2 = x**2

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2, color='green', linewidth=10.0, linestyle='--')


plt.figure(num=3, figsize=(8,5))
plt.plot(x,y2)
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel("I am x")
plt.ylabel("I am y")


plt.figure(num=5, figsize=(8,5))
plt.plot(x,y2)
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel("I am x")
plt.ylabel("I am y")

new_ticks = np.linspace(-1,2,10)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1.5, -1, 1, 3 ],
           ['really bad', 'bad', 'normal', 'good', 'really good'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-1,1,50)
y1 = 2*x+1
y2 = x**2

plt.figure()
plt.plot(x,y1, label='up')
plt.plot(x,y2, color='green', linewidth=10.0, linestyle='--', label='down')
plt.legend()

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#ax.xaxis.set_ticks_position('bottom')
#ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',2.0))
ax.spines['left'].set_position(('data',0.5))


plt.figure(num=3)
l1, = plt.plot(x,y1, label='up')
l2, = plt.plot(x,y2, color='green', linewidth=10.0, linestyle='--', label='down')
plt.legend(handles=[l1,l2,], labels=['aaa','bbb'], loc='best')



plt.figure(num=5)
l1, = plt.plot(x,y1, label='up')
l2, = plt.plot(x,y2, color='green', linewidth=10.0, linestyle='--', label='down')
plt.legend(handles=[l1,], labels=['aaa',], loc='best')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 1024
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
T = np.arctan2(y,x)

plt.scatter(y, x, s=75, c=T, alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()
```
